```python
import pandas as pd
```

```python
netflix_df = pd.read_csv('C:/Users/carad/Desktop/Netflix_Data/netflix1.csv')
```

```python
print(netflix_df)
```

```python
#remove duplicates
netflix_df.drop_duplicates(subset='show_id',
                          keep=False, inplace=True)
```

```python
netflix_df.head()
```

```python
#info on data
netflix_df.info()
```

```python
#change datatype date_added to datetime
netflix_df["date_added"] = pd.to_datetime(netflix_df["date_added"])
netflix_df.info()
```

```python
#add month_added to column
netflix_df['month_added'] = netflix_df['date_added'].dt.month_name()
netflix_df
```

```python
#add year_added column
netflix_df['year_added'] = netflix_df['date_added'].dt.year
netflix_df
```

```python
# added day_added column
netflix_df['day_added']=netflix_df['date_added'].dt.day_name()
netflix_df
```
